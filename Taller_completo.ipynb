# Mejor modelo con forward selection

````python
def forward_selection(X_train, y_train):

    selected_features = []
    remaining_features = list(X_train.columns)
    best_aic = float('inf')

    while remaining_features:
        best_new_feature = None
        improved = False

        for feature in remaining_features:
            current_features = selected_features + [feature]
            X_temp = sm.add_constant(X_train[current_features])
            model = sm.OLS(y_train, X_temp).fit()
            current_aic = model.aic

            if current_aic < best_aic:
                best_aic = current_aic
                best_new_feature = feature
                improved = True

        if improved:
            selected_features.append(best_new_feature)
            remaining_features.remove(best_new_feature)
            print(f"Añadiendo {best_new_feature}. Current AIC: {best_aic:.2f}")
        else:
            break

    return selected_features

selected_features_forward = forward_selection(X_train, y_train)
print("\nVariables seleccionadas usando forward selection:", selected_features_forward)
````

## Mejor modelo usando backward elimination



````python
def backward_elimination(X_train, y_train):

    selected_features = list(X_train.columns)
    X_temp_initial = sm.add_constant(X_train[selected_features])
    model_initial = sm.OLS(y_train, X_temp_initial).fit()
    best_aic = model_initial.aic
    print(f"AIC inicial with all features: {best_aic:.2f}")

    while selected_features:
        feature_to_remove = None
        improved = False
        worst_aic = best_aic # Inicializar con el mejor AIC

        for feature in selected_features:
            temp_features = [f for f in selected_features if f != feature]
            if not temp_features:
                 X_temp = pd.DataFrame({'const': np.ones(len(X_train))})
            else:
                X_temp = sm.add_constant(X_train[temp_features])

            model = sm.OLS(y_train, X_temp).fit()
            current_aic = model.aic

            # En backward elimination, se eliminan las variables que no están, cuando se obtuvo el mejor AIC
            if current_aic < best_aic:
                best_aic = current_aic
                feature_to_remove = feature
                improved = True

        if improved:
            selected_features.remove(feature_to_remove)
            print(f"Variables removidas {feature_to_remove}. Nuevo mejor AIC: {best_aic:.2f}")
        else:
            break

    return selected_features

selected_features_backward = backward_elimination(X_train, y_train)
print("\nVariables seleccionadas usando backward elimination:", selected_features_backward)
````

## Mejor modelo usando stepwise selection




````python
def stepwise_selection(X_train, y_train):

    selected_features = []
    remaining_features = list(X_train.columns)
    best_aic = float('inf')
    no_improvement_count = 0

    while no_improvement_count < 2: # Continue as long as there is potential for improvement (add or remove)
        added_feature = None
        removed_feature = None

        # Step 1: Attempt to add a feature (Forward step)
        best_new_feature = None
        improved_add = False
        current_best_aic_add = best_aic

        for feature in remaining_features:
            current_features = selected_features + [feature]
            X_temp = sm.add_constant(X_train[current_features])
            # Ensure indices are aligned
            X_temp = X_temp.align(y_train, join='inner', axis=0)[0]
            model = sm.OLS(y_train, X_temp).fit()
            current_aic = model.aic

            if current_aic < current_best_aic_add:
                current_best_aic_add = current_aic
                best_new_feature = feature
                improved_add = True

        if improved_add and current_best_aic_add < best_aic:
            best_aic = current_best_aic_add
            selected_features.append(best_new_feature)
            remaining_features.remove(best_new_feature)
            added_feature = best_new_feature
            print(f"Added {best_new_feature}. Current AIC: {best_aic:.2f}")
            no_improvement_count = 0 # Reset counter if improvement made

        # Step 2: Attempt to remove a feature (Backward step)
        feature_to_remove = None
        improved_remove = False
        current_best_aic_remove = best_aic

        if selected_features: # Only attempt removal if there are selected features
             for feature in selected_features:
                temp_features = [f for f in selected_features if f != feature]
                if not temp_features: # Handle the case where removing the last feature
                    X_temp = pd.DataFrame({'const': np.ones(len(X_train))}, index=X_train.index)
                else:
                    X_temp = sm.add_constant(X_train[temp_features])

                # Ensure indices are aligned
                X_temp = X_temp.align(y_train, join='inner', axis=0)[0]
                model = sm.OLS(y_train, X_temp).fit()
                current_aic = model.aic

                if current_aic < current_best_aic_remove:
                    current_best_aic_remove = current_aic
                    feature_to_remove = feature
                    improved_remove = True

        if improved_remove and current_best_aic_remove < best_aic:
            best_aic = current_best_aic_remove
            remaining_features.append(feature_to_remove)
            selected_features.remove(feature_to_remove)
            removed_feature = feature_to_remove
            print(f"Removed {feature_to_remove}. Current AIC: {best_aic:.2f}")
            no_improvement_count = 0 # Reset counter if improvement made

        # Check for improvement in this iteration
        if added_feature is None and removed_feature is None:
            no_improvement_count += 1
        else:
            no_improvement_count = 0


    return selected_features

selected_features_stepwise = stepwise_selection(X_train, y_train)
print("\nSelected features using stepwise selection:", selected_features_stepwise)
````

# Análisis de métricas de desempeño

**RMSE — Root Mean Squared Error**
- Es la raíz del promedio de los errores al cuadrado.
- Penaliza más fuerte los errores grandes (porque se elevan al cuadrado).
- Tiene las mismas unidades que la variable objetivo.
- Más bajo → mejor.
- Útil cuando quieres castigar más los errores grandes.
- Si el RMSE = 103.85, significa que en promedio el modelo se equivoca en sus predicciones en $103.85 (pero con énfasis en errores grandes).

**MAE — Mean Absolute Error**
- Es el promedio de los errores absolutos.
- Penaliza todos los errores por igual.
- Más fácil de interpretar: promedio simple de la magnitud del error.
- Más bajo → mejor.
- Menos sensible a valores atípicos que RMSE.
- Si MAE = 83.33, significa que el modelo se equivos en sus predicciones en promedio $83.33.

**MAPE — Mean Absolute Percentage Error**
- Es el porcentaje promedio de error.
- No depende de las unidades.
- Muy intuitivo para comunicar resultados a gente no técnica.
- Si MAPE = 456%, significa que en promedio el modelo se equivoca un 456% respecto al valor real

  Problemas:
  - No se puede calcular si algún valor de y<sub>i</sub> =0
  - Tiende a inflar el error cuando los valores reales son muy pequeños.

# Regresión con regularizacion

¿Cuándo conviene regularizar?
- Se usa regularización cuando se presenta alguno de estos casos:
- Muchas variables o p ≥ n (alta dimensión). p= variables, n= número de registros.
- Multicolinealidad (predictoras fuertemente correlacionadas).
- Sobreajuste (alto desempeño en entrenamiento, bajo en validación).
- Coeficientes inestables al re-muestrear o pequeños cambios en datos.
- Predicción es prioridad sobre inferencia estricta (p-values/IC).



## Cómo ver las variables seleccionadas?

**Ridge (L2)**

- Ridge nunca pone coeficientes exactamente en 0, solo los “encoge” hacia 0.
- Por tanto, todas las variables permanecen en el modelo, aunque algunas tengan coeficientes muy pequeños (≈0).
- En Ridge no hay selección de variables real, pero puedes ordenar las variables según la magnitud de los coeficientes para ver cuáles son más relevantes.

**Lasso (L1)**

- Lasso sí puede llevar coeficientes exactamente a 0.
- Las variables con βj=0 son descartadas.
- Las que quedan con βj≠0 son las seleccionadas.
- Aquí la selección es explícita: simplemente miras los coeficientes finales.

**Elastic Net (L1 + L2)**
- Combina Lasso y Ridge:
- Puede poner coeficientes en 0 (como Lasso).
- Pero también mantiene grupos de variables correlacionadas (no las descarta todas).
- Variables seleccionadas: aquellas con βj≠0.

````python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Define the features (X) and target (y) based on the previous SFS selection
X_train_regularized = X_train[list(sfs.k_feature_names_)]
X_test_regularized = X_test[list(sfs.k_feature_names_)]

# Ridge Regression
ridge_model = Ridge(alpha=1.0) # You can tune alpha
ridge_model.fit(X_train_regularized, y_train)
y_pred_ridge = ridge_model.predict(X_test_regularized)

print("--- Ridge Regression ---")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_ridge)):.2f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_ridge):.2f}")
mape_ridge = np.mean(np.abs((y_test - y_pred_ridge) / (y_test + epsilon))) * 100
print(f"MAPE: {mape_ridge:.2f}%")
print(f"R2: {r2_score(y_test, y_pred_ridge):.2f}")

# Lasso Regression
lasso_model = Lasso(alpha=1.0) # You can tune alpha
lasso_model.fit(X_train_regularized, y_train)
y_pred_lasso = lasso_model.predict(X_test_regularized)

print("\n--- Lasso Regression ---")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_lasso)):.2f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_lasso):.2f}")
mape_lasso = np.mean(np.abs((y_test - y_pred_lasso) / (y_test + epsilon))) * 100
print(f"MAPE: {mape_lasso:.2f}%")
print(f"R2: {r2_score(y_test, y_pred_lasso):.2f}")

# Elastic Net Regression
elastic_net_model = ElasticNet(alpha=1.0, l1_ratio=0.5) # You can tune alpha and l1_ratio
elastic_net_model.fit(X_train_regularized, y_train)
y_pred_elastic_net = elastic_net_model.predict(X_test_regularized)

print("\n--- Elastic Net Regression ---")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_elastic_net)):.2f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_elastic_net):.2f}")
mape_elastic_net = np.mean(np.abs((y_test - y_pred_elastic_net) / (y_test + epsilon))) * 100
print(f"MAPE: {mape_elastic_net:.2f}%")
print(f"R2: {r2_score(y_test, y_pred_elastic_net):.2f}")
````
